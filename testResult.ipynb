```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data from the line plot
datasets = ['Cricket', 'FaceFour', 'Lightning', 'MoteStrain', 'WindSpeed', 'WindDirection', 'Ecg', 'Wafer', 'Pressure']
bzip2 = [0.3651, 0.3142, 0.3314, 0.3698, 0.2414, 0.2515, 0.386, 0.2951, 0.1276]
Gzip = [0.387, 0.279, 0.352, 0.394, 0.224, 0.225, 0.401, 0.277, 0.126]
TRC = [0.358, 0.294, 0.318, 0.359, 0.252, 0.282, 0.378, 0.303, 0.158]
Our = [0.20955, 0.23876, 0.21648, 0.22622, 0.14673, 0.15336, 0.225, 0.22546, 0.07638]

# Constructing a DataFrame
df = pd.DataFrame({
    'Dataset': datasets,
    'bzip2': bzip2,
    'Gzip': Gzip,
    'TRC': TRC,
    'Our': Our
})

# Convert data to a long-form DataFrame
df_long = pd.melt(df, id_vars='Dataset', var_name='Method', value_name='Compression Ratio')

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 6))

# Set color palette
color_palette = sns.color_palette('tab10')
palette = [color_palette[0], color_palette[2], color_palette[1], color_palette[3]]

# Plot the bars using Seaborn's barplot
bar = sns.barplot(x="Dataset", y="Compression Ratio", hue="Method", data=df_long, palette=palette)

# Customizing the plot to replicate the style of the provided image
ax.set_ylabel(r'$CR$')  # Set the name of the Y-axis
ax.set_title('Compression Ratio when lossless')
ax.legend(loc='upper right')
plt.grid(False)  # Turn off the grid for a cleaner look

# Apply hatch patterns to the bars for "bzip2," "Gzip," and "TRC" across different datasets
hatches = ['/', '\\', 'x']  # Hatch patterns for "bzip2," "Gzip," and "TRC"
hatch_dict = {}
dataset_order = df_long['Dataset'].unique()

for i, dataset in enumerate(dataset_order):
    for j, method in enumerate(["bzip2", "Gzip", "TRC"]):
        hatch_dict[(dataset, method)] = hatches[j]

for i, bar in enumerate(bar.patches):
    dataset = dataset_order[i % len(dataset_order)]  # Cycling through datasets
    method = df_long['Method'][i]
    hatch = hatch_dict.get((dataset, method), "")
    if hatch:
        bar.set_hatch(hatch)

# Save the plot as a PDF file
pdf_filename = '/home/guoyou/ExtractSemantic/figure/crLossless.pdf'
plt.savefig(pdf_filename, format='pdf')

plt.show()

# The path to the saved PDF file
print(pdf_filename)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data from the line plot
datasets = ['Cricket', 'FaceFour', 'Lightning', 'MoteStrain', 'WindSpeed', 'WindDirection', 'Ecg', 'Wafer', 'Pressure']
bzip2 = [0.3651, 0.3142, 0.3314, 0.3698, 0.2414, 0.2515, 0.386, 0.2951, 0.1276]
Gzip = [0.387, 0.279, 0.352, 0.394, 0.224, 0.225, 0.401, 0.277, 0.126]
TRC = [0.358, 0.294, 0.318, 0.359, 0.252, 0.282, 0.378, 0.303, 0.158]
Our = [0.20955, 0.23876, 0.21648, 0.22622, 0.14673, 0.15336, 0.225, 0.22546, 0.07638]

# Constructing a DataFrame
df = pd.DataFrame({
    'Dataset': datasets,
    'bzip2': bzip2,
    'Gzip': Gzip,
    'TRC': TRC,
    'Our': Our
})

# Convert data to a long-form DataFrame
df_long = pd.melt(df, id_vars='Dataset', var_name='Method', value_name='Compression Ratio')

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 6))

# Set color palette and hatches
color_palette = sns.color_palette('tab10')
palette = ['lightgreen', 'pink', 'lightblue', color_palette[3]]
hatches = ['\\', '////', '//////', '']

# Plot the bars using Seaborn's barplot
bar = sns.barplot(x="Dataset", y="Compression Ratio", hue="Method", data=df_long, palette=palette)

# Customizing the plot to replicate the style of the provided image
ax.set_ylabel(r'$CR$')  # Set the name of the Y-axis
ax.set_title('Compression Ratio when lossless')
ax.legend(loc='upper right')
plt.grid(False)  # Turn off the grid for a cleaner look

# Apply hatch patterns to the bars for each method consistently across different datasets
hatch_dict = {method: hatch for method, hatch in zip(df_long['Method'].unique(), hatches)}

for container in bar.containers:
    method = container.get_label()
    hatch = hatch_dict.get(method, "")
    if hatch:
        for bar in container:
            bar.set_hatch(hatch)

# Save the plot as a PDF file
pdf_filename = '/home/guoyou/ExtractSemantic/figure/crLossless.pdf'
plt.savefig(pdf_filename, format='pdf')

plt.show()

# The path to the saved PDF file
print(pdf_filename)
```
